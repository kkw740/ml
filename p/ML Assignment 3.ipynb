```python
data_annual_avg = data[data['ANNUAL'] <= 30.5]
print(data_annual_avg)
print(data_annual_avg.shape)
```

```python
X = data.iloc[:,1:13]
Y = data['ANNUAL']
print(X.shape)
print(Y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
mse = mean_squared_error(y_test,pred)
mae = mean_absolute_error(y_test,pred)
r2 = r2_score(y_test,pred)
print(f"Mean Squared Error(MSE) :{mse}")
print(f"Mean Absolute Error(MSE) :{mae}")
print(f"R2 Score:{r2}")
```

```python
plt.figure(figsize=(8,6))
plt.scatter(y_test,pred,color='red',label='Predicted vs Annual')
plt.plot([min(y_test),max(y_test)],[min(pred),max(pred)],color='blue')
plt.title('Actual vs Predicted Annual Temperature')
plt.xlabel('Actual Annual Temperature')
plt.ylabel('Predicted Annual Temperature')
plt.legend()
plt.show()
```
